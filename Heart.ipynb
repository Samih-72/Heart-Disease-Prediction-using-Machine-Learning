datapreprocessing and data cleaning


```python

df_encode=pd.get_dummies(df,drop_first=True)
```

```python
df_encode 
```

```python
df_encode=df_encode.astype(int)
```

```python
df_encode
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols = [ 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
scaler = StandardScaler()
df_encode[numerical_cols] = scaler.fit_transform(df_encode[numerical_cols])
df_encode.head()
```

feature extraction and scaling

```python
df_encode['AgeGroup'] = pd.cut(
    df_encode['Age'],
    bins=[0,40,55,100],
    labels=['Young','Middle','Senior']
)
```

```python
df_encode
```

```python
pd.get_dummies(df_encode, columns=['AgeGroup'], drop_first=True).astype(int)
```

```python
df_encode['HighCholesterol'] = (
    df['Cholesterol'] > 240
).astype(int)
```

```python
df_encode['LowMaxHR'] = (
    df['MaxHR'] < 100
).astype(int)
```

```python
df_encode
```

```python
df_encode['Age_Cholesterol'] = (
    df['Age'] * df['Cholesterol']
)
```

```python
df_encode['Age_BP'] = (
    df['Age'] * df['RestingBP']
)
```

```python
df_encode['RiskScore'] = (
    (df['Age'] > 60).astype(int) +
    (df['RestingBP'] > 140).astype(int) +
    (df['Cholesterol'] > 240).astype(int) +
    (df['MaxHR'] < 100).astype(int)
)
```

```python
df_encode
```

```python
df_encode['OldpeakRisk'] = (
    df['Oldpeak'] > 2
).astype(int)
```

```python
corr = df_encode.corr(numeric_only=True)

corr['HeartDisease'].sort_values(
    ascending=False
)
```

```python
sns.heatmap(corr,annot=True)
```
